```python
# Load model directly
from transformers import AutoTokenizer, AutoModelForCausalLM

tokenizer = AutoTokenizer.from_pretrained("ibm-granite/granite-3.3-2b-instruct")
model = AutoModelForCausalLM.from_pretrained("ibm-granite/granite-3.3-2b-instruct")
messages = [
    {"role": "user", "content": "Who are you?"},
]
inputs = tokenizer.apply_chat_template(
	messages,
	add_generation_prompt=True,
	tokenize=True,
	return_dict=True,
	return_tensors="pt",
).to(model.device)

outputs = model.generate(**inputs, max_new_tokens=40)
print(tokenizer.decode(outputs[0][inputs["input_ids"].shape[-1]:]))
```

```python
!pip install -q gradio jsonschema pyyaml

import gradio as gr
import json
import yaml
from datetime import datetime
from jsonschema import validate, ValidationError

# ==========================================================
# JSON DATA STUDIO - ULTRA ADVANCED JSON TOOLKIT
# ==========================================================

class JSONStudio:

    def __init__(self):
        self.history = []

    # ------------------------------------------------------
    # Key Value Parser
    # ------------------------------------------------------

    def parse_key_value(self, text):

        data={}
        errors=[]

        for line_num,line in enumerate(text.split("\n"),1):

            line=line.strip()

            if not line:
                continue

            if ":" not in line:
                errors.append(f"Line {line_num} missing ':'")
                continue

            key,value=line.split(":",1)

            key=key.strip()
            value=value.strip()

            data[key]=self.infer_type(value)

        result=json.dumps(data,indent=2)

        self.log("KeyValue→JSON",text,result)

        status="✅ Parsed successfully"

        if errors:
            status+="\n"+ "\n".join(errors)

        return result,status


    # ------------------------------------------------------
    # Type inference
    # ------------------------------------------------------

    def infer_type(self,value):

        if value.lower() in ["true","yes"]:
            return True

        if value.lower() in ["false","no"]:
            return False

        if value.lower() in ["null","none"]:
            return None

        try:
            return int(value)
        except:
            pass

        try:
            return float(value)
        except:
            pass

        return value


    # ------------------------------------------------------
    # JSON → YAML
    # ------------------------------------------------------

    def json_to_yaml(self,json_text):

        try:

            data=json.loads(json_text)

            y=yaml.dump(data,sort_keys=False)

            self.log("JSON→YAML",json_text,y)

            return y,"✅ Converted JSON to YAML"

        except Exception as e:

            return "",str(e)


    # ------------------------------------------------------
    # YAML → JSON
    # ------------------------------------------------------

    def yaml_to_json(self,yaml_text):

        try:

            data=yaml.safe_load(yaml_text)

            j=json.dumps(data,indent=2)

            self.log("YAML→JSON",yaml_text,j)

            return j,"✅ Converted YAML to JSON"

        except Exception as e:

            return "",str(e)


    # ------------------------------------------------------
    # JSON Search
    # ------------------------------------------------------

    def search_json(self,json_text,keyword):

        try:

            data=json.loads(json_text)

            results=[]

            def search(obj,path=""):

                if isinstance(obj,dict):

                    for k,v in obj.items():

                        new_path=f"{path}.{k}" if path else k

                        if keyword.lower() in k.lower():
                            results.append(new_path)

                        search(v,new_path)

                elif isinstance(obj,list):

                    for i,item in enumerate(obj):

                        search(item,f"{path}[{i}]")

                else:

                    if keyword.lower() in str(obj).lower():

                        results.append(path)

            search(data)

            return "\n".join(results),"✅ Search complete"

        except Exception as e:

            return "",str(e)


    # ------------------------------------------------------
    # JSON Statistics
    # ------------------------------------------------------

    def json_stats(self,json_text):

        try:

            data=json.loads(json_text)

            keys=len(data)

            types=set(type(v).__name__ for v in data.values())

            stats=f"""
### 📊 JSON Statistics

**Top Level Keys:** {keys}

**Value Types:**
{types}
"""

            return stats

        except Exception as e:

            return str(e)


    # ------------------------------------------------------
    # JSON Compare
    # ------------------------------------------------------

    def compare_json(self,json1,json2):

        try:

            j1=json.loads(json1)
            j2=json.loads(json2)

            diff=[]

            keys=set(j1.keys()).union(j2.keys())

            for k in keys:

                if j1.get(k)!=j2.get(k):

                    diff.append(f"{k} : {j1.get(k)} -> {j2.get(k)}")

            return "\n".join(diff),"✅ Comparison done"

        except Exception as e:

            return "",str(e)


    # ------------------------------------------------------
    # Fix invalid JSON
    # ------------------------------------------------------

    def auto_fix(self,text):

        try:

            fixed=text.replace("'",'"')

            data=json.loads(fixed)

            result=json.dumps(data,indent=2)

            return result,"✅ Auto fixed"

        except:

            return text,"❌ Unable to fix"


    # ------------------------------------------------------
    # History logger
    # ------------------------------------------------------

    def log(self,action,input_data,output_data):

        self.history.append({

            "time":datetime.now().isoformat(),
            "action":action,
            "input":input_data[:200],
            "output":output_data[:200]

        })


    # ------------------------------------------------------
    # Export history
    # ------------------------------------------------------

    def export_history(self):

        report="JSON DATA STUDIO HISTORY\n\n"

        for i,h in enumerate(self.history,1):

            report+=f"""
Action #{i}
Time: {h['time']}
Type: {h['action']}

Input:
{h['input']}

Output:
{h['output']}

------------------------
"""

        return report


studio=JSONStudio()

# ==========================================================
# DARK THEME CSS
# ==========================================================

dark_css = """
body {
    background: linear-gradient(135deg,#0f172a,#1e293b);
}

/* container */
.gradio-container {
    max-width:1200px !important;
}

/* DARK TITLE PANEL */
h1 {
    background:#020617;
    padding:20px;
    border-radius:12px;
    text-align:center;
    color:#f8fafc !important;
    border:1px solid #334155;
}

/* subtitles */
h2,h3,h4 {
    color:#e2e8f0 !important;
}

/* inputs */
textarea, input {
    background:#1e293b !important;
    color:white !important;
}

/* buttons */
button {
    background:linear-gradient(90deg,#6366f1,#ec4899) !important;
    color:white !important;
    border:none !important;
    border-radius:8px !important;
}
"""

# ==========================================================
# GRADIO UI
# ==========================================================

with gr.Blocks(css=dark_css,title="JSON Data Studio") as demo:

    gr.Markdown("""
# 🚀 JSON Data Studio

### Professional JSON Toolkit

**Features**

- KeyValue → JSON
- JSON ↔ YAML
- JSON Search
- JSON Compare
- Auto Fix JSON
- Statistics Analyzer
- History Tracker
""")

    with gr.Tabs():

        # -------------------------------------
        # KeyValue Parser
        # -------------------------------------

        with gr.TabItem("KeyValue Parser"):

            inp=gr.Textbox(lines=12,label="Input Text")

            btn=gr.Button("Parse")

            out=gr.Code(language="json")

            status=gr.Markdown()

            btn.click(
                studio.parse_key_value,
                inputs=[inp],
                outputs=[out,status]
            )

        # -------------------------------------
        # JSON → YAML
        # -------------------------------------

        with gr.TabItem("JSON → YAML"):

            j=gr.Code(language="json")

            btn=gr.Button("Convert")

            y=gr.Code(language="yaml")

            msg=gr.Markdown()

            btn.click(
                studio.json_to_yaml,
                inputs=[j],
                outputs=[y,msg]
            )

        # -------------------------------------
        # YAML → JSON
        # -------------------------------------

        with gr.TabItem("YAML → JSON"):

            y=gr.Code(language="yaml")

            btn=gr.Button("Convert")

            j=gr.Code(language="json")

            msg=gr.Markdown()

            btn.click(
                studio.yaml_to_json,
                inputs=[y],
                outputs=[j,msg]
            )

        # -------------------------------------
        # JSON Search
        # -------------------------------------

        with gr.TabItem("JSON Search"):

            j=gr.Code(language="json")

            key=gr.Textbox(label="Search keyword")

            btn=gr.Button("Search")

            res=gr.Textbox(lines=10)

            msg=gr.Markdown()

            btn.click(
                studio.search_json,
                inputs=[j,key],
                outputs=[res,msg]
            )

        # -------------------------------------
        # Compare JSON
        # -------------------------------------

        with gr.TabItem("Compare JSON"):

            j1=gr.Code(language="json")
            j2=gr.Code(language="json")

            btn=gr.Button("Compare")

            res=gr.Textbox(lines=10)

            msg=gr.Markdown()

            btn.click(
                studio.compare_json,
                inputs=[j1,j2],
                outputs=[res,msg]
            )

        # -------------------------------------
        # Statistics
        # -------------------------------------

        with gr.TabItem("Statistics"):

            j=gr.Code(language="json")

            btn=gr.Button("Analyze")

            res=gr.Markdown()

            btn.click(
                studio.json_stats,
                inputs=[j],
                outputs=[res]
            )

        # -------------------------------------
        # Auto Fix JSON
        # -------------------------------------

        with gr.TabItem("Auto Fix JSON"):

            t=gr.Textbox(lines=12)

            btn=gr.Button("Fix JSON")

            res=gr.Code(language="json")

            msg=gr.Markdown()

            btn.click(
                studio.auto_fix,
                inputs=[t],
                outputs=[res,msg]
            )

        # -------------------------------------
        # History
        # -------------------------------------

        with gr.TabItem("History"):

            btn=gr.Button("Download History")

            res=gr.Textbox(lines=15)

            btn.click(
                studio.export_history,
                outputs=[res]
            )


demo.launch(share=True)
```
